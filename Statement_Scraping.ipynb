```python
import requests
import pandas as pd
from datetime import datetime as dt
from bs4 import BeautifulSoup
```

```python
load_url = "https://www.federalreserve.gov/monetarypolicy/fomccalendars.htm"
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
url_list = []
elems = soup.select('a')
for elem in elems:
  tmp = elem.get('href')
  try:
    if tmp[-3:] == 'htm':
      if tmp[-5] == 'a':
        print(tmp)
        if tmp[:34] == '/newsevents/pressreleases/monetary':
          url_list.append(tmp)
  except:
    pass
```

```python
len(url_list)
```

*直近5年用*

```python
# 2017年以降
out_dic = {}
for url in url_list:
  load_url = 'https://www.federalreserve.gov' + url
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  if soup.find(class_='title').text == 'FOMC statement' or soup.find(class_='title').text == 'Federal Reserve issues FOMC statement':
    date_as_str = url[-13:-5]
    date_as = dt.strptime(date_as_str, '%Y%m%d')
    out_dic[date_as] = soup.find(class_="col-xs-12 col-sm-8 col-md-8").text
```

```python
df_out = pd.DataFrame(out_dic, index=['text']).T
df_out = df_out.sort_index()
```

2006年～2016年

```python
# 2006～2016
url_list2 = []
for y in range(2006,2018):
  load_url = 'https://www.federalreserve.gov/monetarypolicy/fomchistorical' + str(y) + '.htm'
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  elems = soup.select('a')
  for elem in elems:
    tmp = elem.get('href')
    try:
      if tmp[-3:] == 'htm':
        #if tmp[-5] == 'a':
        if tmp[:34] == '/newsevents/pressreleases/monetary' or tmp[:26] == '/newsevents/press/monetary':
          url_list2.append(tmp)
            #print(tmp)
    except:
      pass
```

```python
out_dic2 = {}
for url in url_list2:
  load_url = 'https://www.federalreserve.gov' + url
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  if soup.find(class_='title').text == 'FOMC statement' or soup.find(class_='title').text == 'Federal Reserve issues FOMC statement':
    date_as_str = url[-13:-5]
    date_as = dt.strptime(date_as_str, '%Y%m%d')
    out_dic2[date_as] = soup.find(class_="col-xs-12 col-sm-8 col-md-8").text
```

```python
df_out2 = pd.DataFrame(out_dic2, index=['text']).T
df_out2 = df_out2.sort_index()
```

1997年～2005年

```python
# 1997～2005
url_list3 = []
for y in range(1990,2006):
  load_url = 'https://www.federalreserve.gov/monetarypolicy/fomchistorical' + str(y) + '.htm'
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  elems = soup.select('a')
  for elem in elems:
    tmp = elem.get('href')
    try:
      if tmp[:25] == '/boarddocs/press/monetary' or tmp[:24] == '/boarddocs/press/general':
        url_list3.append(tmp)
        #print(tmp)
    except:
      pass
```

```python
out_dic3 = {}
for url in url_list3:
  load_url = 'https://www.federalreserve.gov' + url
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  if url[:25] == '/boarddocs/press/monetary':
    date_as_str = url[31:39]
  elif url[:24] == '/boarddocs/press/general':
    date_as_str = url[30:38]
  date_as = dt.strptime(date_as_str, '%Y%m%d')
  out_dic3[date_as] = soup.find('p').text
```

```python
df_out3 = pd.DataFrame(out_dic3, index=['text']).T
df_out3 = df_out3.sort_index()
```

1990年～1997年

```python
# 1990～1997
url_list4 = []
for y in range(1990,1998):
  load_url = 'https://www.federalreserve.gov/monetarypolicy/fomchistorical' + str(y) + '.htm'
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  elems = soup.select('a')
  for elem in elems:
    tmp = elem.get('href')
    try:
      if tmp[:5] == '/fomc':
        if tmp[6] != 'm' and tmp[6] != 'b' and tmp[6] != 'M':
          url_list4.append(tmp)
          #print(tmp)
    except:
      pass
```

```python
out_dic4 = {}
for url in url_list4:
  load_url = 'https://www.federalreserve.gov' + url
  html = requests.get(load_url)
  soup = BeautifulSoup(html.content, "html.parser")
  date_as_str = url[6:14]
  date_as = dt.strptime(date_as_str, '%Y%m%d')
  if date_as_str == '19960131':
    out_dic4[date_as] = soup.find_all('p')[1].text
  else:
    out_dic4[date_as] = soup.find(id="content").text
```

```python
df_out4 = pd.DataFrame(out_dic4, index=['text']).T
df_out4 = df_out4.sort_index()
```

```python
df_out_all = pd.concat([df_out, df_out2, df_out3, df_out4])
df_out_all = df_out_all.sort_index()
```
